# **ðŸ“Š Number of Songs Released Each Year**

```
year_column = 'Year'
year_frequency = data[year_column].value_counts()
year_frequency.sort_values()
```

```
plt.figure(figsize=(10, 6))
plt.bar(year_frequency.index.astype(int), year_frequency.values, color='lightblue')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.title('Song Frequency by Year')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# **ðŸ’¿ Number of Songs in Each Album**

```
Album_column = 'Album'
Album_frequency = data[Album_column].value_counts()
Album_frequency.sort_values()
```

```
plt.figure(figsize=(16, 8))
plt.bar(Album_frequency.index, Album_frequency.values, color='lightblue')
plt.xlabel('Album')
plt.ylabel('Number of Songs')
plt.title('Number of Songs per Album')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# **Sentiment Analysis of Lyrics**

```
data['sentiment'] = data['Lyric'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
print(data[['Title', 'sentiment']].head(10))
```

# **Sentiment Distribution Across Songs**

```
plt.figure(figsize=(8, 5))
plt.hist(data['sentiment'], bins=10, color='skyblue', edgecolor='black')
plt.title('Sentiment Distribution of Coldplay Songs')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Number of Songs')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
